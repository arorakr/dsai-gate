## 1. Check Leap Year

### Problem Statement
Write a Python program that takes a year as input and prints whether it is a leap year or not.

### Example:
Input: 2020  
Output: Leap Year

Input: 1900  
Output: Not a Leap Year

```python
# Write your code here

year = int(input())

if((year%100 == 0) and (year%400 == 0)):
  print("Leap Year")
elif((year%100 != 0) and (year%4 == 0)):
  print("Leap Year")
else:
  print("Not a Leap Year")
```

## 2. Sum of Digits

### Problem Statement
Write a program that takes a number from the user and prints the sum of its digits.

### Example:
Input: 1234  
Output: 10

Input: 567  
Output: 18

```python
# Write your code here
# 5761%10
# 5761 // 10
# sum = 0+1+6+7+5
# num = 0
num = int(input())
sum = 0

while (num != 0):
  last_digit = num % 10
  sum += last_digit

  num //= 10

print(sum)


# str(5761)
# num = str(input())
# sum_ = 0
# for char in num:
#     sum_ += int(char)
# print(sum_)
```

## 3. Palindrome Number

### Problem Statement
Check if the given number is a palindrome or not.

### Example:
Input: 121  
Output: Palindrome

Input: 123  
Output: Not a Palindrome

```python
# Write your code here
# num = str(input())

# if(num == num[::-1]):
#   print("Palindrome")
# else:
#   print("Not a Palindrome")

s = input("Enter a number : ")
r = ""
for char in range(len(s)):
  r = char + r
if r == s:
  print("Palindrome")
else:
  print("Not a Palindrome")
```

## 7. Count Alphabets, Digits, and Special Characters

### Problem Statement
Write a program to count the number of alphabets, digits, and special characters in a string.

### Example:
Input: 'Hello123@#'  
Output: Alphabets = 5, Digits = 3, Special Characters = 2

```python
# Write your code here
string = input()

alpha = 0
digit = 0
special = 0
for char in string:
  if char.isalpha():
    alpha+=1
  elif char.isdigit():
    digit+=1
  else:
    special+=1

print(f"Alphabets = {alpha}, Digit = {digit}, Special Characters = {special}")
```

## 8. Print All Prime Numbers from 1 to N

### Problem Statement
Take an integer N as input and print all prime numbers between 1 and N.

### Example:
Input: 10  
Output: 2 3 5 7

```python
# Write your code here
# Taking input from the user (N)
n = int(input())

# Checking all the integer from 1 to N
for i in range(2,n):
  prime = True
  # Checking current i to be a prime number or not
  for j in range(2,i):
    if i % j == 0:
      prime = False
      break
  if i == 2:
    prime = True

  if prime:
    print(i)
```
